### Realizar consultas del Contenido de la web

```python
# Bloque de web scrapping con resultado de palabras m√°s comunes dentro del texto obtenido

def scrape_website(url, retries=5, backoff_factor=0.3, timeout=10):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    session = requests.Session()
    session.headers.update(headers)

    for i in range(retries):
        try:
            response = session.get(url, timeout=timeout)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')
                paragraphs = soup.find_all('div')
                text = ' '.join([para.get_text() for para in paragraphs])
                # print(text)
                
                with open('resultado_web.txt', 'w', encoding='utf-8') as file:
                    for paragraph in paragraphs:
                        file.write(paragraph.get_text(strip=True) + '\n\n')
                    print('Los p√°rrafos han sido extra√≠dos correctamente y guardados en resultado_web.txt')
                
                return text
            else:
                raise Exception(f"Error al cargar la p√°gina {url}, status code: {response.status_code}")
        except requests.exceptions.RequestException as e:
            print(f"Error: {e}")
            if i < retries - 1:
                sleep_time = backoff_factor * (2 ** i)
                print(f"Reintentando en {sleep_time} segundos...")
                time.sleep(sleep_time)
            else:
                raise Exception(f"Error al intentar conectarse a la p√°gina {url} despu√©s de {retries} intentos")

def get_most_common_words(text, num_words=20):
    # print("Texto antes de limpieza y tokenizaci√≥n:")
    # print(text)

    # Limpieza y tokenizaci√≥n del texto
    text = re.sub(r'\W+', ' ', text).lower()
    words = nltk.word_tokenize(text)
    word_freq = Counter(words)
    common_words = word_freq.most_common(num_words)

    print("Palabras m√°s utilizadas:")
    print(common_words)

    return common_words

def main():
    url = 'http://lobosriders.mywebcommunity.org/'
    scraped_text = scrape_website(url)

    if scraped_text is not None:
        common_words = get_most_common_words(scraped_text, num_words=20)
        print("20 palabras m√°s utilizadas:")
        for word, freq in common_words:
            print(f"{word}: {freq}")
    else:
        print("No se pudo obtener el texto raspado.")

if __name__ == "__main__":
    main()
```

```python
# Realizaci√≥n de tratamiento y limpieza de palabras vac√≠as en Espa√±ol

# Descargar recursos de NLTK si no est√°n presentes
try:
    nltk.data.find('corpora/stopwords.zip')
except LookupError:
    nltk.download('stopwords', quiet=True)

try:
    nltk.data.find('tokenizers/punkt.zip')
except LookupError:
    nltk.download('punkt', quiet=True)

def scrape_website(url, retries=5, backoff_factor=0.3, timeout=10):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    session = requests.Session()
    session.headers.update(headers)

    for i in range(retries):
        try:
            response = session.get(url, timeout=timeout)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')
                paragraphs = soup.find_all('div')
                text = ' '.join([para.get_text() for para in paragraphs])
                #print(text)
                
                with open('resultado2_web.txt', 'w', encoding='utf-8') as file:
                    for paragraph in paragraphs:
                        file.write(paragraph.get_text(strip=True) + '\n\n')
                    print('Los p√°rrafos han sido extra√≠dos correctamente y guardados en resultado2_web.txt')
                
                return text
            else:
                raise Exception(f"Error al cargar la p√°gina {url}, status code: {response.status_code}")
        except requests.exceptions.RequestException as e:
            print(f"Error: {e}")
            if i < retries - 1:
                sleep_time = backoff_factor * (2 ** i)
                print(f"Reintentando en {sleep_time} segundos...")
                time.sleep(sleep_time)
            else:
                raise Exception(f"Error al intentar conectarse a la p√°gina {url} despu√©s de {retries} intentos")

def get_most_common_words(text, num_words=30):
    # Limpieza y tokenizaci√≥n del texto
    text = re.sub(r'\W+', ' ', text).lower()
    words = nltk.word_tokenize(text)
    # Eliminar palabras vac√≠as (stopwords)
    stopwords = set(nltk.corpus.stopwords.words('spanish'))
    words = [word for word in words if word not in stopwords]
    # Filtrar palabras cortas y caracteres no alfab√©ticos
    words = [word for word in words if len(word) > 2 and word.isalpha()]
    word_freq = Counter(words)
    common_words = word_freq.most_common(num_words)
    return common_words

def main():
    url = 'http://lobosriders.mywebcommunity.org/'
    scraped_text = scrape_website(url)

    if scraped_text is not None:
        common_words = get_most_common_words(scraped_text, num_words=20)
        print("20 palabras m√°s utilizadas:")
        for word, freq in common_words:
            print(f"{word}: {freq}")
            
        # Obtener las palabras sugeridas para la metadata
        suggested_words = [word for word, _ in common_words]
        print(f"\nPalabras sugeridas para la metadata: {suggested_words}\n\n\n")
    else:
        print("No se pudo obtener el texto raspado.")

if __name__ == "__main__":
    main()
```
